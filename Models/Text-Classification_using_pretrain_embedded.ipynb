```python
import numpy as np
import pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense

def load_fasttext_embeddings(embedding_file):
    embeddings_index = {}
    with open(embedding_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index

embedding_index = load_fasttext_embeddings('/content/drive/MyDrive/cc.vi.300.vec')
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenized_texts = df_set['free_text'].tolist()
tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(tokenized_texts)

import numpy as np
def prepare_embedding_matrix(tokenizer_word_index, embedding_index, embedding_dim=300):

    number_word = len(tokenizer_word_index) + 1
    embedding_matrix = np.zeros((number_word, embedding_dim))
    for word, idx in tokenizer_word_index.items():
      word = embedding_index.get(word)
      if word is not None:
        embedding_matrix[idx] = word
    return embedding_matrix

embedding_matrix = prepare_embedding_matrix(tokenizer.word_index, embedding_index)
```

```python
sequence_text = tokenizer.texts_to_sequences(tokenized_texts)

data = pad_sequences(sequence_text, maxlen=200, truncating='post', padding='post')
```

```python
labels = df_set['label_id'].tolist()
```

```python
test_data = df_valid['free_text'].tolist()
test_data = tokenizer.texts_to_sequences(test_data)
test_data = pad_sequences(test_data, maxlen=200, truncating='post', padding='post')
test_labels = df_valid['label_id'].tolist()
```

```python
train_feature_tensor =  tf.convert_to_tensor(data)
train_label_tensor =  tf.convert_to_tensor(labels)

test_feature_tensor =  tf.convert_to_tensor(test_data)
test_label_tensor =  tf.convert_to_tensor(test_labels)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((train_feature_tensor, train_label_tensor))
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((test_feature_tensor, test_label_tensor))
```

```python
batch_size_train = 32
batch_size_test = 16
shuffle_buffer = 1000
```

```python
def process_dataset(dataset, batch_size, shuffle_buffer):
    ds = dataset.shuffle(shuffle_buffer)
    return ds.batch(batch_size).prefetch(1)
```

```python
train_dataset = process_dataset(train_dataset, batch_size_train, shuffle_buffer)
```

```python
test_dataset = process_dataset(test_dataset, batch_size_test, shuffle_buffer)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, RNN, Dense, LSTM, Bidirectional, SimpleRNN
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam, RMSprop, SGD
```

```python
vocab_size = len(list(tokenizer.word_index)) + 1
embedding_size = 300
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_size, weights=[embedding_matrix], trainable = True, input_length = 200))
model.add(Bidirectional(LSTM(16, return_sequences=True)))
model.add(Bidirectional(LSTM(16, return_sequences=True)))
model.add(LSTM(8))
model.add(Dense(4, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(16, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="adam", loss='binary_crossentropy', metrics=['acc'])
```

```python
model.fit(train_dataset, epochs=5, validation_data = test_dataset)
```
